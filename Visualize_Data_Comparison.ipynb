```python
# prompt: ดรอป na

movie_file_path = movie_file_path.dropna(subset=['Rating','Duration','Genre']) # Changed 'Ratng' to 'Rating'
movie_file_path = movie_file_path[['Rating','Duration','Genre']] # Changed 'Ratng' to 'Rating'\
movie_file_path.head()
```

```python
# prompt: ลบ min แล้วก็เปลี่ยนประเภทข้อมูลขอลความยาว เป็น int

movie_file_path['Duration'] = movie_file_path['Duration'].str.replace('min', '')
movie_file_path['Duration'] = movie_file_path['Duration'].astype(int)
```

```python
movie_file_path.describe()
```

```python
# prompt: แยกตารางออกเป็นตารางย่ิย 5 ตารางสำหรับ genre:Drama,Action ,horror.musical

# Create a dictionary to store the dataframes for each genre
genre_dfs = {}

# Define the genres to extract
genres = ['Drama', 'Action', 'Horror', 'Musical','Romance']

# Loop through the genres and create separate dataframes
for genre in genres:
  genre_dfs[genre] = movie_file_path[movie_file_path['Genre'] == genre]

# Access the dataframes using the genre as the key
# Example:
drama_df = genre_dfs['Drama']
action_df = genre_dfs['Action']
horror_df = genre_dfs['Horror']
musical_df = genre_dfs['Musical']
Romance_df = genre_dfs['Romance']
```

```python
alltable = pd.concat(genre_dfs.values())
alltable.head()
```

```python
# prompt: แบ่งความยาว ออกเป็น shot,mid ,long  โดย short สั้นกว่า 60 mid 60-90ที่เหลือเป็นยาว

# Create a new column 'Duration_Category' to store the duration category
movie_file_path['Duration_Category'] = ''

# Assign duration categories based on the criteria
alltable.loc[movie_file_path['Duration'] < 60, 'Duration_Category'] = 'short'
alltable.loc[(movie_file_path['Duration'] >= 60) & (movie_file_path['Duration'] <= 90), 'Duration_Category'] = 'mid'
alltable.loc[movie_file_path['Duration'] > 90, 'Duration_Category'] = 'long'
alltable.head()
```

```python
india_bar_data = alltable.groupby(['Genre','Duration_Category'])['Rating'].mean().unstack()
india_bar_data
```

```python
ax = india_bar_data.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Genre')
plt.ylabel('Rating')
plt.title('Rating by Genre and Duration Category')

plt.legend(bbox_to_anchor=(1.05, 1), loc='center left', borderaxespad=0.)

plt.gca().get_legend().remove()

plt.show()
```

```python
# prompt: จากกราฟข้างบน เพิ่มป้ายบอกให้หน่อยว่าสีไหนคืออะไร

ax = india_bar_data.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Genre')
plt.ylabel('Rating')
plt.title('Rating by Genre and Duration Category')

# Add labels and title
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.title('Rating by Genre and Duration Category')

# Move legend outside the chart
plt.legend(bbox_to_anchor=(1.05, 1), loc='center left', borderaxespad=0.)

# Add labels to the bars
for p in ax.patches:
    ax.annotate(str(round(p.get_height(), 2)), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Show the chart
plt.show()
```

## Radar Chart

```python
import numpy as np
import pandas as pd  #import pandas มาเพราะข้อมูลจะเกี่ยวกับตาราง
import matplotlib.pyplot as plt
from math import pi #import math มาเนื่องจากเกี่ยวกับวงกลมนำมาใช้เพื่อดึงค่าพาย
```

```python
data = {'client_id' : ['c1' , 'c2' , 'c3' , 'c4'],
         'cheese' : [2,5,4,3],
         'beer' : [2,1,3,5],
         'chicken' : [5,3,3,4],
        'bread' : [4,3,2,1],
       'coffee' : [2,3,5,3]}
df = pd.DataFrame.from_dict(data)
df
```

```python
categories = list(df.columns[1:])
categories
```

```python
values = df.mean().values.flatten().tolist()
values += values[:1] # repeat the first value to close the circular graph
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='grey', size=12)
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

plt.ylim(0, 5)
ax.set_rlabel_position(30)

ax.plot(angles, values, linewidth=1, linestyle='solid' , color = '#ea3033') #เส้นกราฟ
ax.fill(angles, values, 'lightpink', alpha=0.5) #พื้นที่
```

```python
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='grey', size=12)
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

plt.ylim(0, 5)
ax.set_rlabel_position(30)

# part 1
val_c1 = df.loc[0].drop('client_id').values.flatten().tolist()
val_c1 += val_c1[:1]
ax.plot(angles, val_c1, linewidth=1,
        linestyle='solid', label='Client c1')
ax.fill(angles, val_c1, 'skyblue', alpha=0.4)

# part 2
val_c2=df.loc[1].drop('client_id').values.flatten().tolist()
val_c2 += val_c2[:1]
ax.plot(angles, val_c2, linewidth=1,
        linestyle='solid', label='Client c2')
ax.fill(angles, val_c2, 'lightpink', alpha=0.4)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
```
