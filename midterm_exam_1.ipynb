```python
import os

path = '/content/drive/MyDrive/Data_viz2024_Data/Test'

test_TidyData = os.path.join(path,'test-TidyData_2024.xlsx')
print(test_TidyData )
```

```python
test_TidyData = pd.read_excel(test_TidyData)
test_TidyData
```

```python
test_TidyData.info()
```

```python
test_TidyData.shape
```

```python
# แยกปีออกจากคอลัมน์ month
test_TidyData['year'] = test_TidyData['month'].str.split('_').str[1]
test_TidyData
```

```python
# สร้าง DataFrame สำหรับปี 2018 และ 2019
data_2018 = test_TidyData[test_TidyData['year'] == '2018']
data_2019 = test_TidyData[test_TidyData['year'] == '2019']

```

```python
# คำนวณค่าเฉลี่ยของ Chlo_a สำหรับแต่ละ station ในปี 2018 และ 2019
mean_2018 = data_2018.groupby('station')['Chlo_a'].mean()
mean_2019 = data_2019.groupby('station')['Chlo_a'].mean()
```

```python
# หาความแตกต่างของค่าเฉลี่ย
difference = mean_2019 - mean_2018

# แสดงผลลัพธ์
print(difference)
```

```python
# สร้างตาราง Pivot Table เพื่อคำนวณค่าเฉลี่ย Chlo_b สำหรับแต่ละ station และเดือน
chlo_b_mean_table = pd.pivot_table(test_TidyData, values='Chlo_b', index='station', columns='month', aggfunc='mean')

# แสดงผลลัพธ์
print(chlo_b_mean_table)
```
