````
# prompt: ลบ html tags และ html codes ออกจาก text ใน column description ใช้ beautifulsoup

from bs4 import BeautifulSoup

def remove_html_tags(text):
  soup = BeautifulSoup(text, "html.parser")
  cleaned_text = soup.get_text()
  return cleaned_text

folk_df['description'] = folk_df['description'].apply(remove_html_tags)
````

````
# prompt: ลบ \xa0 และ \n ออกไปด้วย

def remove_special_chars(text):
  cleaned_text = text.replace('\xa0', '')
  cleaned_text = cleaned_text.replace('\n', '')
  return cleaned_text

folk_df['description'] = folk_df['description'].apply(remove_special_chars)
````

1.วนลูบหาคำว่าแม่หรือพ่อกี่เรื่อง เรื่องอะไรบ้าง และคนแต่ง

2.มีคำดังกล่าวใน นิทานชีวิต หรือนิทานประจำถิ่นมากกว่ากัน

3.คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่อง ว่า แม่ หรือ พ่อ มากที่สุด

````
#ข้อ 1
count_mom_dad = 0
list_story = []
for each_row in this_data.iterrows():
    if ('แม่' in each_row[1]['title']) or ('พ่อ' in each_row[1]['title']):
        count_mom_dad += 1
        list_story.append([each_row[1]['title'],each_row[1]['creator']])

print(f'มีคำว่า พ่อ หรือ แม่ ในชื่อเรื่อง {count_mom_dad} เรื่อง')
print(f'รายชื่อนิทานดังกล่าว')
for i in list_story:
    print(i)

#ข้อ 2
count_life = 0
count_local = 0
for each_row in this_data.iterrows():
    if ('แม่' in each_row[1]['title']) or ('พ่อ' in each_row[1]['title']):
        if each_row[1]['type'] == 'นิทานชีวิต':
            count_life += 1
        elif each_row[1]['type'] == 'นิทานประจำถิ่น':
            count_local += 1

if count_life > count_local:
    print('นิทานชีวิต มีคำว่า แม่ หรือ พ่อ มากกว่า')
elif count_life < count_local:
    print('นิทานประจำถิ่น มีคำว่า แม่ หรือ พ่อ มากกว่า')
else:
    print('นิทานชีวิต และ นิทานประจำถิ่น มีคำว่า แม่ หรือ พ่อ เท่ากัน')

#ข้อ 3
creator_count = {}
for each_row in this_data.iterrows():
    if ('แม่' in each_row[1]['title']) or ('พ่อ' in each_row[1]['title']):
        creator = each_row[1]['creator']
        if creator not in creator_count:
            creator_count[creator] = 0
        creator_count[creator] += 1

max_count = 0
max_creator = ''
for creator, count in creator_count.items():
    if count > max_count:
        max_count = count
        max_creator = creator

print(f'คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุดคือ {max_creator} แต่งไป {max_count} เรื่อง')
````
