````
# จัดการค่า float ที่อาจเกิดขึ้นในคอลัมน์ 'description'
folk_df['description_length'] = folk_df['description'].apply(lambda x: len(str(x)) if not isinstance(x, float) else 0)
````

````
longest_tale = folk_df.loc[folk_df['description_length'].idxmax()]
shortest_tale = folk_df.loc[folk_df['description_length'].idxmin()]

print("นิทานที่ยาวที่สุด:")
print(longest_tale[['title', 'description_length']])

print("นิทานที่สั้นที่สุด:")
print(shortest_tale[['title', 'description_length']])
````

````
# ชนิดของนิทานที่มากที่สุดและน้อยที่สุด
type_counts = folk_df['type'].value_counts()
most_common_type = type_counts.idxmax()
least_common_type = type_counts.idxmin()

print("ชนิดของนิทานที่มีจำนวนมากที่สุด:")
print(most_common_type, type_counts[most_common_type])

print("ชนิดของนิทานที่มีจำนวนน้อยที่สุด:")
print(least_common_type, type_counts[least_common_type])
````

````
# นิทานที่มี spatial ที่ไม่ใช่ไทยกี่เรื่อง
non_thai_spatial_count = folk_df[folk_df['spatial'] != 'ไทย'].shape[0]

print("จำนวนเรื่องที่ spatial ไม่ใช่ไทย:")
print(non_thai_spatial_count)
````

### การวนลูป record ในตาราง  (```.iterrows```)

````
this_data = data_covid[['sex','age','province_of_onset']]
````

### การวนลูป แบบมองตารางแพนด้าส์(pandas dataframe) เป็น numpy array  (```.iloc```)

````
for each_row in range(this_data.shape[0]):
    if (this_data.iloc[each_row,1] == 20) and (this_data.iloc[each_row,2] == 'ขอนแก่น'):  #each_row คือ ตัวเลข
        print(each_row)
        print(this_data.iloc[each_row,:])
````
